```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Menggunakan double backslash
df = pd.read_csv("C:\\Users\\Lenovo\\Downloads\\archive (1).zip")
```

```python
# Display first few rows of the dataset
df.head()
```

```python
# Convert 'Bulan' to datetime format and extract the year
df['Bulan'] = pd.to_datetime(df['Bulan'])
df['Tahun'] = df['Bulan'].dt.year
```

```python
# Display the first few rows again to check the changes
df.head()
```

```python
# Group data by 'Tahun' and sum 'Penjualan'
penjualan_tahunan = df.groupby('Tahun')['Penjualan'].sum().reset_index()
```

```python
# Plot yearly sales
plt.figure(figsize=(10, 6))
sns.lineplot(data=penjualan_tahunan, x='Tahun', y='Penjualan', marker='o')
plt.title('Tren Penjualan Mobil per Tahun')
plt.xlabel('Tahun')
plt.ylabel('Total Penjualan')
plt.grid(True)
plt.show()
```

```python
# Extract month name and group by month to find average sales per month
df['Bulan_nama'] = df['Bulan'].dt.month_name()
penjualan_bulanan = df.groupby('Bulan_nama')['Penjualan'].mean().reindex(
    ["January", "February", "March", "April", "May", "June", 
     "July", "August", "September", "October", "November", "December"]
).reset_index()
```

```python
# Display the monthly sales table
print(penjualan_bulanan)
```

```python
# Plot average sales by month
plt.figure(figsize=(12, 6))
sns.barplot(data=penjualan_bulanan, x='Bulan_nama', y='Penjualan')
plt.title('Rata-rata Penjualan Mobil per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Rata-rata Penjualan')
plt.xticks(rotation=45)
plt.show()
```

```python

# Save the processed dataset to a specific folder
df.to_csv("C:/Users/Lenovo/Downloads/penjualan_mobil_processed.csv", index=False)
```
